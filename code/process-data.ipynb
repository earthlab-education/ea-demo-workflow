```python
# Import needed packages
import os
import os.path as op
import matplotlib.pyplot as plt

# Importing custom package
import list_generator
```

```python
# Creating a directory to store the new image in
images_dir_path = op.join(op.dirname(op.abspath('')), 'images')

if not op.isdir(images_dir_path):
    os.mkdir(images_dir_path)
```

```python
# Getting a list of integers from the custom package
data_to_plot = list_generator.random_list()

# Plotting list
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(data_to_plot)
ax.set(title="Fun Data!", xlabel="Variable 1", ylabel="Variable2")

plt.show()
```

```python
# Saving the image to the images directory
fig.savefig(op.join(images_dir_path, 'line_plot.png'), bbox_inches='tight', dpi=150)
```

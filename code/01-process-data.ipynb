## Custom Imports

Below you can see that the custom module called list_generator.py is imported
directly. Following PEP8 standards it should be in a separate "group" of 
imports separated by a space.

Note that the import below works because this notebook is in the same directory
as the .py file. It's not easy to import the module without that

```python
import os
import os.path as op

import matplotlib.pyplot as plt

# Importing custom package
import list_generator
```

```python
# Create directory to store new image
images_dir_path = op.join(op.dirname(op.abspath('')), 'images')

# Check that is exists, and if it doesn't, create it
if not op.isdir(images_dir_path):
    os.mkdir(images_dir_path)
```

```python
# Generate list of integers using the custom module - list_generator
data_to_plot = list_generator.random_list()

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(data_to_plot)
ax.set(title="Fun Data!",
       xlabel="Variable 1",
       ylabel="Variable2")

plt.show()
```

```python
# Save the image to the images directory
fig.savefig(op.join(images_dir_path, 'line_plot.png'),
            bbox_inches='tight', dpi=150)
```
